```python
import numpy as np
import cv2 as cv
```

```python
from google.colab.patches import cv2_imshow
```

```python
img = cv.imread('/content/sample_data/step1_original-frame.png')
```

```python
output_img = img.copy()
```

```python
cv2_imshow(output_img)
```

```python
gray_img = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
```

```python
blured_gray_img = cv.medianBlur(gray_img, 5)
```

```python
circles = cv.HoughCircles(blured_gray_img,cv.HOUGH_GRADIENT,1,204,param1=50,param2=30,minRadius=0,maxRadius=0)
```

```python
detected_circles = np.uint16(np.around(circles)) #convert circles into int array (x,y of center & r length)
```

```python
for (x, y, r) in detected_circles[0, :]:
  cv.circle(output_img, (x,y), r, (255,0,0), 2) #draw the border of the detected circle
  cv.circle(output_img, (x,y), 2, (0,255,255), 3) #draw marker on the center of r=2
```

```python
cv2_imshow(output_img)
```
